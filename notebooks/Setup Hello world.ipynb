```python
from sentence_transformers import SentenceTransformer

#1. Load a pretrained Sentence Transformers model
model=SentenceTransformer("all-MiniLM-L6-v2")

```

```python
import numpy as np
```

```python
#The idea is suppose we have a document like this
doc=["Hi This is Akshar and i'm an associate software enginner",
     "I have been work as a full stack python developer",
    ]

#Now we want to convert this doc into a numerical representation i.e vector embeddings
```

```python
#We can calculate the embeddings using model.encode()
embeddings=model.encode(doc)
print(f"This is the vector embedding{embeddings}")
print(embeddings.shape)  #(2,384)
```

```python
print(doc)
splitted_docs=doc[0].split()
print(splitted_docs)
```

```python
test1_doc=["The dog jumped over the cat"]
test2_doc=["The cat jumped over the dog"]

```

```python
test1_embeddings=model.encode(test1_doc)
test2_embeddings=model.encode(test2_doc)

#Now we will compare to actually see how similar they are 
```

```python
#Lets see similarity between same embeddings
model.similarity(test1_embeddings,test1_embeddings)

#It will give tensor(1) that means they are exactly similar
```

```python
#Lets see similarity between test1_embeddings and test2_embeddings
model.similarity(test1_embeddings,test2_embeddings)

#It gave 98 percent match
```

```python
#Suppose we have a document and a query and we have to give the similarity of the query to the document
document=["mera naam hai akshar","mera dost hai akshat"]
embedding=model.encode(document)
print(embedding)
embedding.shape
query="my name is akshar"
```

```python
model.similarity(embedding, model.encode(query))
```
